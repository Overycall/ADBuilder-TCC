```python
import pandas as pd

df_150 = pd.read_csv('ADBuilder_Dataset_150_APKs.csv')
df_250 = pd.read_csv('ADBuilder_Dataset_250_APKs.csv')
df_350 = pd.read_csv('ADBuilder_Dataset_350_APKs.csv')

print("df_150_APKs shape: ", df_150.shape)
print("df_250_APKs shape: ", df_250.shape)
print("df_350_APKs shape: ", df_350.shape)
```

```python
df_150.shape[1]
```

```python
import matplotlib.pyplot as plt

# Dados
dados = {
    '150': (df_150.shape[0], df_150.shape[1]),
    '250': (df_250.shape[0], df_250.shape[1]),
    '350': (df_350.shape[0], df_350.shape[1])
}

# Separação dos dados
x_labels = list(dados.keys())
y1_values = [d[0] for d in dados.values()]
y2_values = [d[1] for d in dados.values()]

# Configuração do gráfico
fig, ax = plt.subplots()
bar = ax.bar(x_labels, y2_values)

# Personalização do gráfico
ax.set_xlabel('APKs')
ax.set_ylabel('Qtd. de Características')
ax.legend()

# Adicionando os valores nas barras
for bar in bar:
    height = bar.get_height()
    ax.annotate('{}'.format(height),
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),
                textcoords="offset points",
                ha='center',
                va='bottom')

# Exibição do gráfico moderno
plt.show()
```
